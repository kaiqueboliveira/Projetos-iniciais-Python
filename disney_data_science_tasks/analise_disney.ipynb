## IMPORTAR BIBLIOTECAS NECESSARIAS

```python
# importando a biblioteca
from bs4 import BeautifulSoup as bs
import requests
```

## CARREGAR A PAGINA DA WEB

```python
# realizando o acesso
r = requests.get("https://en.wikipedia.org/wiki/Toy_Story_3")

# converteção para o objeto beautifulSoup
soup = bs(r.content)


# imprimir o HTML
contents = soup.prettify()
print(contents)
```

```python
#Obtem todas a linhas de uma tabela HTMl
info_box = soup.find(class_ = "infobox vevent")
info_rows = info_box.find_all("tr")
for row in info_rows:
    print(row.prettify())
```

```python
#extração das informações e tratamento da tabela HTML.
def get_content_value(row_data):
    if row_data.find("li"):
        return [li.get_text().replace('\xa0',' ') for li in row_data.find_all("li")]
    else:
        return row.find("td").get_text().replace('\xa0', ' ')



#construção do dicionario associando cada detalhe a uma chave.
movie_info = {}
for index, row in enumerate(info_rows):
    if index == 0:
        movie_info['title'] = row.find("th")
    elif index == 1:
         continue
    else:
        content_key = row.find("th").get_text()
        content_value = get_content_value(row.find("td"))
        movie_info[content_key] = content_value

movie_info

```
